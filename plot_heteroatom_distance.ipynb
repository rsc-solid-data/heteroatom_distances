```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("H1_distance_Group_1.csv", delimiter=",", skiprows=1)
x = data[:, 0]
y = data[:, 1]

print(f"mean bond length: {np.mean(x):.3f} Ã…")

plt.vlines(np.mean(x), 0, np.max(y), colors='r', linestyles='dashed')

plt.plot(x, y)
plt.title("Gaussian Distribution")
plt.xlabel("Distance (Ã…)")
plt.ylabel("Probability Density")
plt.show()
```
